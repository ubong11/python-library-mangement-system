```python
# Library list
library = []

# Adding books
library.append("Things Fall Apart")
library.append("Rich Dad Poor Dad")
library.append("The Alchemist")

print("Library:", library)
```

```python
print("\nList of Books:")

for index, book in enumerate(library):
    print(f"{index}: {book}")
```

```python
def check_availability(book_name):
    if book_name in library:
        return f"{book_name} is available."
    else:
        return f"{book_name} is not available."

# Test
print(check_availability("The Alchemist"))
print(check_availability("Harry Potter"))
```

```python
def read_books_from_file(filename):
    try:
        with open(filename, "r") as file:
            books = file.readlines()
            return [book.strip() for book in books]
    except FileNotFoundError:
        print("File not found.")
        return []

# Create sample file
with open("books.txt", "w") as f:
    f.write("Book A\nBook B\nBook C")

print(read_books_from_file("books.txt"))
```

```python
import pickle

def save_books_pickle(filename, data):
    try:
        with open(filename, "wb") as file:
            pickle.dump(data, file)
        print("Books saved successfully.")
    except Exception as e:
        print("Error saving:", e)


def load_books_pickle(filename):
    try:
        with open(filename, "rb") as file:
            return pickle.load(file)
    except FileNotFoundError:
        print("Pickle file not found.")
        return []

# Test
save_books_pickle("library.pkl", library)
loaded_books = load_books_pickle("library.pkl")

print("Loaded books:", loaded_books)
```

```python
import sqlite3

def add_books_to_db(books):
    try:
        conn = sqlite3.connect("library.db")
        cursor = conn.cursor()

        cursor.execute("""
        CREATE TABLE IF NOT EXISTS books (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            title TEXT
        )
        """)

        for book in books:
            cursor.execute("INSERT INTO books (title) VALUES (?)", (book,))

        conn.commit()

        # Fetch data
        cursor.execute("SELECT * FROM books")
        rows = cursor.fetchall()

        print("\nBooks in Database:")
        for row in rows:
            print(row)

        conn.close()

    except Exception as e:
        print("Database error:", e)

# Test
add_books_to_db(library)
```

```python
def remove_book(book_name):
    if book_name in library:
        library.remove(book_name)
        print(f"{book_name} removed.")
    else:
        print("Book not found.")

remove_book("Rich Dad Poor Dad")
print(library)
```

```python
def update_book(old_name, new_name):
    if old_name in library:
        index = library.index(old_name)
        library[index] = new_name
        print("Book updated.")
    else:
        print("Book not found.")

update_book("The Alchemist", "The Alchemist - Updated")
print(library)
```
